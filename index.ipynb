## Loading the JSON file

Load the data from the file `disease_data.json`.

```python
#Your code here 
import json
f = open('disease_data.json')
data = json.load(f)
```

## Explore the first and second levels of the schema hierarchy

```python
#Your code here
type(data['meta'])
data['meta']['view'].keys()
```

```python
for i in data['meta']['view'].keys():
    print(i, type(i))
```

```python
display(data['meta']['view']['columns'])
```

## Convert to a DataFrame

Create a DataFrame from the JSON file. Be sure to retrive the column names for the dataframe. (Search within the 'meta' key of the master dictionary.) The DataFrame should include all 42 columns.

```python
#Your code here
cols = []
for i in range(42):
    cols.append(data['meta']['view']['columns'][i]['name'])

import pandas as pd
pd.set_option('display.max_columns', 999)

df = pd.DataFrame(data['data'],columns = cols)
display(df)
```

## Level-Up
## Create a bar graph of states with the highest asthma rates for adults age 18+

```python
df_asthma = df.loc[df['Question'] == "Current asthma prevalence among adults aged >= 18 years"]

df_asthma.DataValue.unique()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


df_asthma['DataValue'].fillna(value = '0.0', inplace = True)

df_asthma.DataValue.unique()
```

```python
import numpy as np

plt.figure(figsize =(18,12))
plt.xticks(rotation=90)
plt.yticks(range(0,150,10))
plt.title("Asthma Rates for Adults 18+ By State")
plt.bar(df_asthma.LocationAbbr, df_asthma.DataValue, )
plt.show()
```
